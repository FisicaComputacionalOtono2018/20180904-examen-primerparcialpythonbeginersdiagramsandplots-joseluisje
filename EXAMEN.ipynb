```
#2
import math
import matplotlib.pyplot as plt
x=[-100,]
d=-100
while d<=100:
  x.append(d)
  d=d+0.1
ye=[]
for i in x:
  a=-(i**2)+3*i
  ye.append(a)
r=float(3)/2
print("El punto en el que cambia la curva es: x=",r)
#La segunda derivada de la función es f'(x)=-2x+3 y sabemos por el criterio o teorema de la segunda derivada que si f'(a)=0, entonces f tienen un máxmio o mínimo relativo en a.
#f'(x)=-2x+3 si y solo si -2x+3=0 si y solo si x=-3/2, por lo que f'(x) tiene un cambio de signo en el punto x=-3/2, es decir, f(x) tiene un cambio de curva en el punto x=-3/2
plt.plot(x,ye)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```
#3
print("Sea f(x) un polinomio de la forma: \n             f(x)= ax2+ bx + c \nIntroducir los valores de los coeficientes a, b y c, así como un valor real de x para evaluar el polinomio en este punto.")
a=int(input("Primer coeficiente: "))
b=int(input("Segundo coeficiente: "))
c=int(input("Tercer coeficiente: "))
d=int(input("Valor de X: "))
e=d**2
h=a*e+b*d+c
print("El valor del polinomio evaluado en X es: ",h)
x=[-2,]
f=-2
while f<=4:
  f=f+0.1
  x.append(f)
ye=[]
for i in x:
  j=i**2
  k=a*j+b*i+c
  ye.append(k)
print("La gráfica del polinomio f(x) en el rango [-2,4] es: ")
plt.plot(x,ye)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
